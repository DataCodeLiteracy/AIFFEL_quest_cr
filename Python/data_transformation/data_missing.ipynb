```python
import pandas as pd
import os
```

```python
salary1_data_path = os.getenv("HOME") + "/programming/AI/AIFFEL/AIFFEL_quest_cr/Python/data/salary_1.csv"
salary2_data_path = os.getenv("HOME") + "/programming/AI/AIFFEL/AIFFEL_quest_cr/Python/data/salary_2.csv"
cpi_path = os.getenv("HOME") + "/programming/AI/AIFFEL/AIFFEL_quest_cr/Python/data/cpi.csv"
job_path = os.getenv("HOME") + "/programming/AI/AIFFEL/AIFFEL_quest_cr/Python/data/job.csv"
```

```python
salary1_data = pd.read_csv(salary1_data_path)
salary2_data = pd.read_csv(salary2_data_path)
cpi_data = pd.read_csv(cpi_path)
job_data = pd.read_csv(job_path)
```

```python
salary1_data.head()
salary2_data.head()
```

```python
salary_df = pd.concat([salary1_data, salary2_data])
salary_df.loc[0] # 중복되는 인덱스가 존재한채로 병합된다.
```

```python
salary_df.reset_index(drop = True, inplace = True) # drop -> 빠져나온 인덱스 제거, inplace -> 덮어쓰기
salary_df
```

```python
# cpi 데이터 확인
cpi_data.head()
salary_df['Country'].unique()
cpi_data['Country'].unique()
```

```python
# 국가 이름 간소화
cpi_data['Country'] = cpi_data['Country'].replace({'United States': 'USA', 'United Kingdom': 'UK'}) 
cpi_data
```

```python
salary_df = salary_df.merge(cpi_data, on = 'Country', how = 'left')
salary_df.drop(['Reference', 'Previous', 'Units', 'Frequency'], axis = 1, inplace = True)
salary_df.rename({'Last': 'CPI'}, axis = 1, inplace = True)
salary_df
```

```python
# 하지만.. CPI값이 object로 병합됨.. 
salary_df.info()
```

```python
salary_df['CPI'] = pd.to_numeric(salary_df['CPI'])
```

```python
salary_df.info()
salary_df.describe()
```

```python
# 결측치 찾기
salary_df.isna().sum()
```

```python
salary_df.isna().mean()
```

```python
# 연차 데이터가 들쑥날쑥이라 크게 결측값을 대체하기 보다 삭제해도 상관 없을 데이터
salary_df[salary_df['Age'].isna()]
```

```python
salary_df = salary_df.dropna()
salary_df.isna().sum()
```

```python
# describe()로 이상치 확인해서 수정 및 삭제
salary_df.describe()
```

```python
# min 값에서 -1이 이상 데이터라고 판단
salary_df = salary_df[~(salary_df['Years of Experience'] < 0)]
salary_df.describe()
```

```python
# 나이에 맞는 연차 데이터 확인 후 수정
salary_df = salary_df[~(salary_df['Years of Experience'] > salary_df['Age'] - 15)]
# 그로 인해 말도 안되는 82년 연차라는 데이터는 제외됨
salary_df.describe()
```

```python
# 이렇게 알고 싶은 그룹에 컬럼 값의 평균을 알아내도 되지만, 유용한 메서드를 활용해서 간략하게 사용하자.
salary_df[salary_df['Gender'] == 'Male']['Salary'].mean()
salary_df[salary_df['Gender'] == 'Female']['Salary'].mean()
```

```python
salary_df.groupby('Gender')['Salary'].mean()
salary_df.groupby('Gender')['Salary'].sum()
salary_df.groupby('Gender')['Salary'].median()
salary_df.groupby('Gender')['Salary'].std()
```

```python
# 숫자 타입 컬럼의 평균 값들 구하기
numeric_cols = salary_df.select_dtypes(include='number')
result = salary_df[['Gender']].join(numeric_cols).groupby('Gender').mean()
result
```

```python
salary_df.groupby(['Gender', 'Country'])['Salary'].mean()
salary_df.groupby(['Gender', 'Country'])['Salary'].agg(['sum', 'mean'])
salary_df.groupby(['Gender', 'Country'])['Salary'].mean().reset_index()
```
