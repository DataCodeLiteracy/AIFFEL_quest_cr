```python
import pandas as pd
import os
import numpy as np
```

```python
salary1_data_path = os.getenv("HOME") + "/programming/AI/AIFFEL/AIFFEL_quest_cr/Python/data/salary_1.csv"
salary2_data_path = os.getenv("HOME") + "/programming/AI/AIFFEL/AIFFEL_quest_cr/Python/data/salary_2.csv"
```

```python
salary1_data = pd.read_csv(salary1_data_path)
salary2_data = pd.read_csv(salary2_data_path)
salary_df = pd.concat([salary1_data, salary2_data])
salary_df.reset_index(drop = True, inplace = True) # drop -> 빠져나온 인덱스 제거, inplace -> 덮어쓰기
salary_df
```

```python
pd.pivot_table(salary_df, index = 'Gender', columns = 'Country', values = 'Salary')
pd.pivot_table(salary_df, index = 'Gender', columns = 'Country', values = 'Salary', aggfunc = np.mean)
pd.pivot_table(salary_df, index = 'Gender', columns = 'Country', values = 'Salary', aggfunc = 'sum')
pd.pivot_table(salary_df, index = ['Gender', 'Race'], columns = 'Country', values = 'Salary', aggfunc = 'sum')
```

```python
sales_df = pd.DataFrame({'company': ['a','a','a','a','b','b','b','b'], 
                    'quarter': ['q1','q2','q3','q4','q1','q2','q3','q4'],
                    'sales': [111,222,333,444,555,666,777,888]})

sales_temp = pd.pivot(sales_df, index = 'company', columns = 'quarter', values = 'sales')
# sales_temp.columns = sales_temp_columns.rename('') # 동작 안됨..
sales_temp.columns.name = ''
new_sales_df = sales_temp.reset_index()
new_sales_df = pd.melt(new_sales_df, id_vars = 'company', value_vars = ['q1','q2','q3','q4'], value_name = 'sales')
new_sales_df
```
